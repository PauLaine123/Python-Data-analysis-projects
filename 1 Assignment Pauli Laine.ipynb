```python
import pandas as pd
import numpy as np
```

```python
#Getting the dataset
HR_dataset = pd.read_csv('HR_Analytics_Dataset.csv')
```

```python
#Investigating data
HR_dataset.head(20)
```

```python
#The average number of years employees work before leaving the company
Filtered_dataset = HR_dataset[HR_dataset['left_company'] == 'Yes'][['years_at_company']]
np.average(Filtered_dataset['years_at_company'])
```

```python
#Percentage of the employees a promotion in the last 5 years

Promoted_employees = HR_dataset[HR_dataset['promotion_last_5_years'] == 'Yes'].shape[0]
total_employees = HR_dataset.size
percentage_of_promoted = (Promoted_employees / total_employees) * 100
print(f"Percentage of people promoted in the last 5 years is: {percentage_of_promoted:.2f}%")
```

```python
#Comparing the promotion rate for employees who left vs employees who stayed
Promoted_employees_who_left = HR_dataset[(HR_dataset['promotion_last_5_years'] == 'Yes') & (HR_dataset['left_company'] == 'Yes')].shape[0]
Promoted_employees_who_stayed = HR_dataset[(HR_dataset['promotion_last_5_years'] == 'Yes') & (HR_dataset['left_company'] == 'No')].shape[0]
total_employees = HR_dataset.size
percentage_of_promoted_who_left = (Promoted_employees_who_left / total_employees) * 100
percentage_of_promoted_who_stayed = (Promoted_employees_who_stayed / total_employees) * 100
print(f"Promoted people who have stayed is: {percentage_of_promoted_who_stayed:.2f}%")
print(f"Promoted people who have left is: {percentage_of_promoted_who_left:.2f}%")
```

```python
#Top 2 departments with the highest attrition rates
Attrition_rate = (HR_dataset[HR_dataset['left_company'] == 'No'].groupby('department').size() / HR_dataset.size)
Top2_departments_by_attrition = (Attrition_rate.nlargest(2) * 100).apply(lambda x: f"{x:.2f}%")
Top2_departments_by_attrition
```

```python
#My suggestions for HR policies to address high attrition:
    #Based on previous analysis, we can see that IT and Marketing departments have the highest attributions, so the main focus should be on other departments.
    #Using the code below, we can see that HR department is the lowest, followed by Finance and Sales, so necessary policies should be directed there.
Departments_attrition_rate = HR_dataset[HR_dataset['left_company'] == 'No'].groupby('department').size() / HR_dataset.size
Departments_attrition_rate_percent = (Departments_attrition_rate * 100).apply(lambda x: f"{x:.2f}%")
Departments_attrition_rate_percent
```

```python
#Operations analytics
```

```python
Operations_dataset = pd.read_csv('Operations_Analytics_Dataset.csv')
```

```python
Operations_dataset.head(20)
```

```python
#What percentage of deliveries were late?
Late_deliveries = (Operations_dataset[Operations_dataset['on_time_delivery'] == 'No']).size
Late_deliveries_percent = ((Late_deliveries / Operations_dataset.size) * 100)
print(f"Late deliveries percentage: {Late_deliveries_percent:.2f}%")
```

```python
#The product category with the highest number of late deliveries
late_deliveries = Operations_dataset[Operations_dataset['on_time_delivery'] == 'No']
late_deliveries_by_category = late_deliveries.groupby('product_category')['on_time_delivery'].count()
top_category = late_deliveries_by_category.nlargest(1)
print(f"Product category with the highest late deliveries:{top_category}")

```

```python
#Estimated potential savings per warehouse when 10% reducement
#Savings annually = 0.1x * 50000
late_deliveries = Operations_dataset[Operations_dataset['on_time_delivery'] == 'No']
late_deliveries_by_warehouse = late_deliveries.groupby('warehouse_location').size()
saving_per = late_deliveries_by_warehouse * 0.10
total_late_deliveries = late_deliveries_by_warehouse.sum()
savings_per_warehouse = (saving_per / total_late_deliveries) * 50000
savings_per_warehouse
```

```python
#Suggest one operational improvement based on analysis
#Based on the analysis with Pandas, company could cut costs in warehouses located in the east and west, since most of the deliveries are late in those parts.
#Company should take more action to reduce the late deliveries in the clothing category, such as improving logistical operations and customer delivery.
```

```python
Financial_dataset = pd.read_csv('Financial_Analytics_Dataset.csv')
```

```python
Financial_dataset.head(20)
```

```python
#Replace missing values in sales_amount & profit_margin with mean
Filtered_columns = Financial_dataset[['sales_amount', 'profit_margin']]
Mean_value = Filtered_columns.mean() 
Replacing_null_columns = Filtered_columns.fillna(Mean_value)
Replacing_null_columns
```

```python
#Number of transactions meeting the following criteria:
    #Profit margin greater than 20%
    #Sales Amount greater than 300
    #Transaction occurred in 2023-07

Filtered_transactions = Financial_dataset[(Financial_dataset['profit_margin'] > 20) & (Financial_dataset['sales_amount'] > 300) & (Financial_dataset['transaction_month'] == '2023-07')]
Result = Filtered_transactions.shape[0]
print(f"Amount of transactions over 20% profit margin, 300 sales amount and registered in 2023-07 is: {Result}")
```

```python
#New profit_amount column with: (sales_amount * profit margin /) 100
Financial_dataset['profit_amount'] = (Financial_dataset['sales_amount'] * Financial_dataset['profit_margin']) / 100
Financial_dataset[['profit_amount', 'sales_amount', 'profit_margin']]
```

```python
#Top 5 transactions with the highest profit amount
Financial_dataset[['transaction_id', 'profit_amount']].nlargest(5, 'profit_amount')
#Transactions 291, 117, 732, 1358 and 670 are the top 5 profitable transactions
```
